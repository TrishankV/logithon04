```python
import pandas as pd
import numpy as np
import os
import PyPDF2 as pp
```

```python
def split_pdf(input_path, output_folder):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the input PDF file
    with open(input_path, 'rb') as file:
        pdf_reader = pp.PdfReader(file)
    # Use getNumPages instead of len() for older PyPDF2 versions
        num_pages = len(pdf_reader.pages)

        # Iterate through each page
        for page_number in range(num_pages):
            pdf_writer = pp.PdfWriter()
      # Access page using indexing
            pdf_writer.add_page(pdf_reader.pages[page_number])

            # Output PDF file name
            output_filename = os.path.join(output_folder, f'page_{page_number + 1}.pdf')

            # Write the page to a new PDF file
            with open(output_filename, 'wb') as output_file:
                pdf_writer.write(output_file)
    return num_pages

                
def split_multiple_pdfs(pdf_paths, output_folder):
    # Iterate over each PDF path
    n = []
    for pdf_path in pdf_paths:
        # Split the PDF
        num = split_pdf(pdf_path, output_folder)
        n.append(num)
    return n
```
